```python
#create file handel

filename = "excelsheet.csv"

file = open(filename, "r")


#write data in file
fileContent = file.read()
#save and close file
file.close()

rows = fileContent.split("\n")
for row in rows:
    print(f"row is{row}")
    cols = row.split(",")
    for col in cols:
        print(col)
```
